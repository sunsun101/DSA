```python
class Node:
    def __init__(self, val):
        self.l = None
        self.r = None
        self.v = val

class Tree:
    def __init__(self):
        self.root = None

    def getRoot(self):
        return self.root

    def add(self, val):
        if self.root is None:
            self.root = Node(val)
        else:
            self._add(val, self.root)

    def _add(self, val, node):
        if val < node.v:
            if node.l is not None:
                self._add(val, node.l)
            else:
                node.l = Node(val)
        else:
            if node.r is not None:
                self._add(val, node.r)
            else:
                node.r = Node(val)

    def find(self, val):
        if self.root is not None:
            return self._find(val, self.root)
        else:
            return None

    def _find(self, val, node):
        if val == node.v:
            return node
        elif (val < node.v and node.l is not None):
            return self._find(val, node.l)
        elif (val > node.v and node.r is not None):
            return self._find(val, node.r)

    def deleteTree(self):
        self.root = None

    def printTree(self):
        if self.root is not None:
            self._printTree(self.root)

    def _printTree(self, node):
        if node is not None:
            self._printTree(node.l)
            print(str(node.v) + ' ')
            self._printTree(node.r)
            
    # For finding successor and deletion operation      
    def findParent(self, node, x):
        if (node.l and node.l.v == x.v) or (node.r and node.r.v == x.v):
            return node
        else:
            if node.l is not None:
                return self.findParent(node.l , x)
            if node.r is not None:
                return self.findParent(node.r , x)
            
    def getTreeMinimum(self, node):
        while node.l is not None:
            node = node.l
        return node
        
    def findSuccessor(self, val):
        node = self.find(val)
        if node is None:
            print("Given node is not found in the tree")
            return None
        if node.r is not None:
            return self.getTreeMinimum(node.r).v
        y = self.findParent(self.root, node)
        while y is not None and node == y.r:
            node = y
            y = self.findParent(self.root,y)
        return y.v if y else None
    
    def transplant(self,u,v):
        if self.findParent(self.root, u) is None:
            self.root = v
        elif u == self.findParent(self.root, u).l:
            self.findParent(self.root, u).l = v
        else:
            self.findParent(self.root, u).r = v
        if v is not None:
            v_parent = self.findParent(self.root, v)
            u_parent = self.findParent(self.root, u)
            v_parent = u_parent
            
    
    def deleteNode(self, val):
        node = self.find(val)
        if node is None:
            print("Given node is not found in the tree")
            return
        if node.l is None:
            # case - one right child
            self.transplant(node, node.r)
        elif node.r is None:
            #case - one left child
            self.transplant(node, node.l)
        else:
            y = self.getTreeMinimum(node.r)
            # case: successor is not direct child
            y_parent = self.findParent(self.root, y)
            if y_parent != node:
                print("successor is not the direct child")
                self.transplant(y , y.r)
                y.r = node.r
                y_r_parent = self.findParent(self.root, y.r)
                y_r_parent = y
            self.transplant(node, y)
            y.l = node.l
            y_l_parent = self.findParent(self.root, y.l)
            y_l_parent = y
                
            

tree = Tree()
tree.add(3)
tree.add(4)
tree.add(0)
tree.add(8)
tree.add(2)
tree.printTree()
print(tree.find(3).v)
print(tree.find(10))
#For finding successor
print("Printing Successors")
print("Successor of 3: ",tree.findSuccessor(3))
print("Successor of 4: ",tree.findSuccessor(4))
print("Successor of 0: ",tree.findSuccessor(0))
print("Successor of 8: ",tree.findSuccessor(8))
print("Successor of 2: ",tree.findSuccessor(2))
print("Successor of 10: ",tree.findSuccessor(10))

print("Deleting 3")
tree.deleteNode(3)
tree.printTree()

print("Deleting 2")
tree.deleteNode(2)
tree.printTree()

tree.deleteTree()
tree.printTree()
```
